Here's an explanation of the code used to split the `image_array_colored` into its individual Red, Green, and Blue channels:

### **Red Channel Extraction**

```python
red_channel = image_array_colored.copy()
red_channel[:,:,1]=0
red_channel[:,:,2]=0
display(red_channel)
```

*   `red_channel = image_array_colored.copy()`: This line creates a copy of the original color image NumPy array. It's crucial to use `.copy()` to ensure that modifications to `red_channel` don't affect the original `image_array_colored`.
*   `red_channel[:,:,1]=0`: In an RGB image array, the second channel (index 1) typically represents Green. By setting all pixel values in this channel to 0, we effectively remove the green component from the image.
*   `red_channel[:,:,2]=0`: Similarly, the third channel (index 2) represents Blue. Setting these values to 0 removes the blue component.
*   `display(red_channel)`: This displays the resulting image, which now primarily contains only the red color information, as green and blue components have been zeroed out.

### **Green Channel Extraction**

```python
green_channel =image_array_colored.copy()
green_channel[:,:,0]=0
green_channel[:,:,2]=0
display(green_channel)
```

*   This process is analogous to the red channel extraction. A copy of the original image is made. The Red channel (index 0) and Blue channel (index 2) are set to 0, leaving only the Green channel's information visible.

### **Blue Channel Extraction**

```python
blue_channel =image_array_colored.copy()
blue_channel[:,:,0]=0
blue_channel[:,:,1]=0
display(blue_channel)
```

*   Again, a copy is made, and this time the Red channel (index 0) and Green channel (index 1) are set to 0, isolating the Blue channel.

 Principal Component Analysis (PCA)

````python
import numpy as np
import matplotlib . pyplot as plt
from PIL import Image
# Load image and convert to grayscale
image = Image . open ("/content/drive/MyDrive/level6/ai/week1/cameraman.png") . convert ("L")
display ( image )
# Convert to NumPy array
image_array = np . array ( image )
# Get image dimensions
height , width = image_array . shape
# Reshape the image into a 2D array where each row represents a row of pixels
# This allows PCA to process each row as an observation and find patterns across pixel intensities
data = image_array . reshape ( height , width )
````

````python
# Compute the mean of each column ( feature )
mean = np .mean ( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean
````

````python
# Compute covariance matrix
cov_matrix = np . cov ( centered_data , rowvar = False )
````

````python
# Compute eigenvalues and eigenvectors
eigenvalues , eigenvectors = np . linalg . eigh ( cov_matrix )
# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np . argsort ( eigenvalues ) [:: -1]
eigenvalues = eigenvalues [ sorted_indices ]
eigenvectors = eigenvectors [: , sorted_indices ]
````

````python
 # Plot the explained variance ratio
explained_variance_ratio = eigenvalues / np .sum ( eigenvalues )
plt . plot ( np . cumsum ( explained_variance_ratio ) )
plt . title (" Cumulative Explained Variance ")
plt . xlabel (" Number of Components ")
plt . ylabel (" Cumulative Explained Variance ")
plt . grid ( True )
plt . show ()
````

````python
k = 50 # Choose k principal components
components = eigenvectors [: , : k ]
````

````python
compressed_data = np . dot ( centered_data , components )
````

````python
decompressed_data = np .dot ( compressed_data , components .T ) +mean
````

````python
import matplotlib.pyplot as plt

# Assuming you already have:
# image_array       -> original grayscale image as 2D NumPy array
# compressed_data   -> compressed representation (e.g., SVD components)
# decompressed_data -> reconstructed image from compression
# k                 -> number of components used in compression

plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Compressed Representation
plt.subplot(1, 3, 2)
plt.imshow(compressed_data, cmap="gray", aspect="auto")
plt.title(f"Compressed Image (k={k})")
plt.axis("off")

# Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap="gray")
plt.title("Decompressed Image")
plt.axis("off")

plt.tight_layout()
plt.show()
````
