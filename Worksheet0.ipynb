```python
def length():
    """
    Converts length between meters and feet.

    Parameters:
        Takes user input for unit type and numeric value.

    Returns:
        Prints the converted value.
    """
    try:
        unit = input("Enter unit (m for meters, ft for feet): ").lower()
        value = float(input("Enter the value to convert: "))

        if unit == "m":
            result = value * 3.28084
            print("Meters to Feet:", result, "ft")

        elif unit == "ft":
            result = value / 3.28084
            print("Feet to Meters:", result, "m")

        else:
            print("Unsupported length unit.")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


def weight():
    """
    Converts weight between kilograms and pounds.

    Parameters:
        Takes user input for unit type and numeric value.

    Returns:
        Prints the converted value.
    """
    try:
        unit = input("Enter unit (kg for kilograms, lbs for pounds): ").lower()
        value = float(input("Enter the value to convert: "))

        if unit == "kg":
            result = value * 2.20462
            print("Kilograms to Pounds:", result, "lbs")

        elif unit == "lbs":
            result = value / 2.20462
            print("Pounds to Kilograms:", result, "kg")

        else:
            print("Unsupported weight unit.")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


def volume():
    """
    Converts volume between liters and gallons.

    Parameters:
        Takes user input for unit type and numeric value.

    Returns:
        Prints the converted value.
    """
    try:
        unit = input("Enter unit (L for liters, gal for gallons): ").lower()
        value = float(input("Enter the value to convert: "))

        if unit == "l":
            result = value * 0.264172
            print("Liters to Gallons:", result, "gal")

        elif unit == "gal":
            result = value / 0.264172
            print("Gallons to Liters:", result, "L")

        else:
            print("Unsupported volume unit.")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


def conv():
    """
    Main function that asks the user which type of conversion they want.

    Parameters:
        None

    Returns:
        Calls the appropriate conversion function.
    """
    conversion = input("Choose conversion type (length / weight / volume): ").lower()

    if conversion == "length":
        length()
    elif conversion == "weight":
        weight()
    elif conversion == "volume":
        volume()
    else:
        print("Invalid conversion type.")


# Run the program
conv()
```

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
        numbers (list): list of numeric values

    Returns:
        float: sum of the numbers
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
        numbers (list): list of numeric values

    Returns:
        float: average of the numbers
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Finds the maximum value in a list.

    Parameters:
        numbers (list): list of numeric values

    Returns:
        float: maximum number
    """
    return max(numbers)


def find_min(numbers):
    """
    Finds the minimum value in a list.

    Parameters:
        numbers (list): list of numeric values

    Returns:
        float: minimum number
    """
    return min(numbers)


def main():
    """
    Main function to select an operation and perform it
    on a list of numbers entered by the user.
    """
    operation = input("Choose operation (sum / average / max / min): ").lower()

    try:
        numbers = input("Enter numbers separated by space: ").split()
        numbers = [float(num) for num in numbers]

        if len(numbers) == 0:
            print("Error: List cannot be empty.")
            return

        if operation == "sum":
            print("Sum =", find_sum(numbers))

        elif operation == "average":
            print("Average =", find_average(numbers))

        elif operation == "max":
            print("Maximum =", find_max(numbers))

        elif operation == "min":
            print("Minimum =", find_min(numbers))

        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Error: Please enter valid numeric values.")


# Run program
main()
```

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): A nested list containing sublists.

    Returns:
        list: A single flattened list containing all elements.
    """
    flat_list = []

    for sublist in lst:
        for item in sublist:
            flat_list.append(item)

    return flat_list


# Example
numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)

print("Original list:", numbers)
print("Flattened list:", result)
```

```python
def access_nested_element(lst, indices):
    """
    Returns a specific element from a nested list using given indices.

    Parameters:
        lst (list): The nested list.
        indices (list): A list of indices that specify the position.

    Returns:
        element: The value located at the specified indices.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# Example
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(lst, indices)

print("Nested List:", lst)
print("Accessed Element:", result)
```

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    Parameters:
        lst (list): A nested list that may contain numbers or other lists.

    Returns:
        int/float: The total sum of all numeric elements in the nested list.
    """
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)   # recursive call
        else:
            total += item

    return total


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = sum_nested(numbers)

print("Nested list:", numbers)
print("Sum of all elements:", result)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): The nested list.
        elem: The element to remove.

    Returns:
        list: A new nested list with the element removed.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result


# Example
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

result = remove_element(lst, elem)

print("Original list:", lst)
print("Modified list:", result)
```

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
        lst (list): A nested list that may contain numbers or other lists.

    Returns:
        int/float: The maximum element in the nested list.
    """
    maximum = None

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)   # recursive call
        else:
            value = item

        if maximum is None or value > maximum:
            maximum = value

    return maximum


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = find_max(numbers)

print("Nested list:", numbers)
print("Maximum element:", result)
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
        lst (list): The nested list.
        elem: The element to count.

    Returns:
        int: Number of occurrences of elem in the nested list.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)   # recursive call
        elif item == elem:
            count += 1

    return count


# Example
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

result = count_occurrences(lst, elem)

print("Nested list:", lst)
print("Occurrences of", elem, ":", result)
```

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
        lst (list): A nested list that may contain lists within lists.

    Returns:
        list: A flattened list containing all elements.
    """
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive call
        else:
            flat_list.append(item)

    return flat_list


# Example
lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

result = deep_flatten(lst)

print("Original list:", lst)
print("Flattened list:", result)
```

```python
def average_nested(lst):
    """
    Returns the average of all numbers in a nested list.
    """
    total = 0
    count = 0

    for item in lst:
        if isinstance(item, list):
            for num in item:
                total += num
                count += 1
        else:
            total += item
            count += 1

    return total / count


# Example
lst = [[1, 2], [3, 4], [5, 6]]

result = average_nested(lst)

print("Nested list:", lst)
print("Average:", result)
```
